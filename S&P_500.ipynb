Data Cleaning and Visualization

```python
import matplotlib.pyplot as plt
```

```python
df.set_index("Date", inplace = True)
df.index = pd.to_datetime(df.index)
```

```python
df.Close.astype('float').dtypes
```

```python
plt.figure(figsize = (20,10))
plt.plot(df["Close"])
plt.title("S&P 500 Daily",fontsize = 20)
plt.ylabel("S&P 500 Index Close",fontsize = 15)
plt.xlabel("Date",fontsize = 15)
plt.grid()
plt.show()
```

In the year of 2022, the S&P 500 Index has been constantly declining with fluctuations. The Index peaked at the beginning of 2022 and had it lowest point in October.
